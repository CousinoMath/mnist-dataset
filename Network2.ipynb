# Network2 #

## Preliminaries ##

```python
%pip install mnist

import numpy as np
import gzip
import _pickle
import src.nndl.network2 as network2
import matplotlib.pyplot as plt

with gzip.open('./data/mnist.pkl.gz', 'rb') as fp:
    training_inputs, training_results, testing_inputs, testing_results = \
        _pickle.load(fp)

training_data = [(ti, tr)
                 for ti, tr in zip(training_inputs, training_results)]
testing_data = [(ti, tr)
                for ti, tr in zip(testing_inputs, testing_results)]
networks = []
monitor = []
```

## Network 1 `[784, 30, 10]` ##

```python
networks.append(network2.Network([784, 30, 10]))
net = networks[0]
epochs = 100

monitor.append(
    net.stochastic_gradient_descent(training_data, epochs, 10, 0.01,
                                     lambda_ = 5.0,
                                     evaluation_data=testing_data,
                                     monitor_training_cost=True,
                                     monitor_training_accuracy=True,
                                     monitor_evaluation_cost=True,
                                     monitor_evaluation_accuracy=True))
```

```python
(testing_cost, testing_accuracy, training_cost, training_accuracy) = \
    monitor[0]
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(np.arange(0, epochs),
       [accuracy / 100.0 for accuracy in testing_accuracy],
       color='#2a6ea6',
       label="Accuracy on test data")
plt.plot(np.arange(0, epochs),
        [accuracy * 100.0 / 60000 for accuracy in training_accuracy],
        color='#ffa933',
        label="Accuracy on training data")
plt.grid(True)
plt.xlabel("Epoch")
plt.ylabel("% Correct")
plt.legend(loc="lower right")
plt.show()

plt.subplot(1, 2, 2)
plt.plot(np.arange(0, epochs),
        [cost for cost in testing_cost],
        color='#2a6ea6',
        label="Cost on test data")
plt.plot(np.arange(0, epochs),
        [cost for cost in training_cost],
        color='#ffa933',
        label="Cost on training data")
plt.grid(True)
plt.xlabel("Epoch")
plt.ylabel("Cost (cross-entropy)")
plt.legend(loc="lower right")
plt.show()
```
